```python
ml100kid_2_plot_vector = pd.read_pickle("ml100kid_2_plot_vector.pkl")
```

```python
import numpy as np
```

```python
def filminfo2features(f_id,film_info):
#    cast_block = [0] * len(filtered_cast)
    cast_block = []
#    directors_block = [0] * len(filtered_directors)
    directors_block = []
    genres_block = [0] * len(all_genres)
#     for c in film_info['cast']:
#         if c in filtered_cast:
#             cast_block[filtered_cast.index(c)] = 1
#     for d in film_info['director']:
#         if d in filtered_directors:
#             directors_block[filtered_directors.index(d)] = 1
    for g in film_info['genres']:
        if g in all_genres:
            genres_block[all_genres.index(g)] = 1
            
    
    features = cast_block + directors_block + genres_block + list(ml100kid_2_plot_vector[f_id])
    for ind, f in enumerate(features):
        if f == np.nan or f == np.inf or str(f) == "nan":
            features[ind] = 0
    return features
```

```python
filminfo2features(16552, ml100k_id_2_meta_info[16552])
```

```python
X = []
y = []

for i in ml100k_id_2_meta_info:
    if str(i) in ml100k_id_2_netflix_vector:
        X.append(filminfo2features(int(i),ml100k_id_2_meta_info[i]))
        y.append(ml100k_id_2_netflix_vector[str(i)])
```

```python
print(len(X))
print(len(y))
print(len(X[0]))
print(len(y[0]))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
```

```python
from sklearn.neural_network import MLPRegressor

reg = MLPRegressor(max_iter =50000, hidden_layer_sizes = (1000, 500, 200), random_state = 42, learning_rate_init = 0.001, verbose = True, early_stopping = True, tol=0.1).fit(X_train, y_train)

reg.score(X_train, y_train)
```

```python
#BEST RES
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.decomposition import PCA

reg = RandomForestRegressor(max_samples = 0.8, n_estimators = 100, random_state = 32, max_depth = 500, max_features = 0.9, n_jobs = -1, verbose = True)

estimators = [('reduce_dim', PCA(n_components = 300)), ('clf', reg)]
pipe = Pipeline(estimators)

pipe.fit(X_train, y_train)

pipe.score(X_train, y_train)
```

```python
reg = MLPRegressor(max_iter =50000, batch_size = 2, hidden_layer_sizes = (500), random_state = 42, learning_rate_init = 0.00001, verbose = True, early_stopping = True, tol=0.00000001)

estimators = [('reduce_dim', PCA(n_components = 300)), ('clf', reg)]
pipe = Pipeline(estimators)

pipe.fit(X_train, y_train)

pipe.score(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```
