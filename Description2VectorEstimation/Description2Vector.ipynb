```
from nltk.corpus import stopwords
eng_stopwords = stopwords.words('english')
```

```
plot_corpus = []
for k in ml100k_id_2_meta_info:
    plot_corpus.append(" ".join(list(list([lemmatizer.lemmatize(x) for x in list(filter(lambda tok: tok not in eng_stopwords ,tokenizer.tokenize(" ".join(ml100k_id_2_meta_info[k]['plot']).lower())))]))))
```

```
def get_plot2vector(plot):
  #plot: lst of str
  plot = " ".join(plot)
  tfidf_res = tfidf_vectorizer.transform([plot]).toarray()[0]
  temp = tfidf_res.argsort()
  ranks = np.arange(len(tfidf_res))[temp.argsort()]
  max_rank = max(ranks)
  vec = [0] * 300
  n = 0
  for rank in range(5):
    ind = np.where(ranks == max_rank - rank )[0][0]
    word = tfidf_vectorizer.get_feature_names()[ind]
    try:
      vec += pretrained_word_vects[word]
      n+=1
    except KeyError:
      pass
  return np.array(vec) / n

```
